### B.2.1 For two given lists list1 and list2 of integers with equal length define the following new lists:

1. *list_zip* which combines the two lists to a single list like a zip. So start with the first element of list1, followed by the first element of *list2*, than the second elements until all elements of the lists are combined.


```python
list = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
list_zip = []

for i in range(len(list)):
    list_zip.append(list[i])
    list_zip.append(list2[i])

print(list_zip)
```

2. *list_odd* which contains only the odd integers of *list1*, followed by the odd integers of *list2*.


```python
list = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
list_odd = []

for i in range(len(list)):
    if list[i] % 2 != 0:
        list_odd.append(list[i])

for i in range(len(list2)):
    if list2[i] % 2 != 0:
        list_odd.append(list2[i])

print(list_odd)
```

3. *list_zip_reverse* which combines the elements of the lists like for *list_zip* but starts at the last element of *list1*.

```python
list = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
list_reverse = []

for i in range(len(list)):
    list_reverse.append(list[len(list)-1-i])
    list_reverse.append(list2[len(list2)-1-i])
    

print(list_reverse)
```

## B.4 Difference between assignment, copy and deep copy

### B.4.1 Have a look at the module copy and the functions copy, deepcopy as well as the standard assignment operator =.

Run the following code and explain its output:

```python
import copy
list1 = [[1, 2, 3], [4, 5, 6], ["a", "b", "c"], 2]
list2 = list1
list3 = copy.copy(list1)
list4 = copy.deepcopy(list1)

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list1[3] = -2

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list2[2][2] = 9

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list1.append([0, 8, 15])

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)
```

1. = assigns a alias -> list2 points at the same object as list1, thus the same ID and the same behaviour.
2. Copy creates a shallow copy of the content with same ID for the elements of the inner list but new ID for the outer list. The inner elements are being changed with list1 but the outer elements stay the same.
3. Deepcopy creates a complete clone of the List. Neither the inner nor the outer lists have the same IDs, thus stays the same throughout the modifications of other lists.

To write this in Python follow these steps:

1. Search for a module that allows to generate random values in python.
2. Define a function *def in_unit_circle(N)* that computes and returns *M* from the function input *N* , which is a single positive integer. Hint: You can interpret two numbers between 0 and 1 as cartesian coordinates of a point and the squared sum of them will tell you the distance of this point from the origin.
3. Define a function *def estimate_pi(N)* that computes and returns the estimation of $\pi$ with the above equation.
4. Search for a module that provides the exact value of $\pi$ and write a function def get_accuracy(N) that returns the absolute difference between the above function and $\pi$.
5. Test your function with different values of *N*. **Hint**: *N* needs to be quite large to have multiple digits of $\pi$ correct.

```python
import random
import math

def in_unit_circle(N): #Generates N random points in the unit square and counts how many M are in the unit circle
    M = 0
    for _ in range(N):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            M += 1
    return M
#print(in_unit_circle(1000))

def estimate_pi(N): #Estimates pi by using the in_unit_circle function
    M = in_unit_circle(N)
    pi_estimate = 4 * M / N
    return pi_estimate
#print(estimate_pi(1000000))

def get_accuracy(N): #Returns the absolute difference between the estimated pi and the actual pi
    pi_estimate = estimate_pi(N)
    accuracy = abs(math.pi - pi_estimate)
    return accuracy

testlist = [10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000]
for N in testlist:
    print(f"N: {N}, Estimated pi: {estimate_pi(N)}, Accuracy: {get_accuracy(N)}")
```

```python
import random
import math

def in_unit_circle(N: int) -> int: #Generates N random points in the unit square and counts how many M are in the unit circle
    M = 0
    if N <= 0:
        raise ValueError("N must be a positive integer")
    for _ in range(N):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            M += 1
    return M
#print(in_unit_circle(1000))

def estimate_pi(N: int) -> float: #Estimates pi by using the in_unit_circle function
    if N <= 0:
        raise ValueError("N must be a positive integer")
    M = in_unit_circle(N)
    pi_estimate = 4 * M / N
    return pi_estimate
#print(estimate_pi(1000000))
```

N: int ist the type hint for function variables, -> int describes the return type. N<0 and Value Error secures the int is positive and not 0.

```python
from typing import List

def fib_first(n: int) -> List[int]:
   
    if n < 0:
        raise ValueError("n must be non-negative")

    def rec(k: int, a: int, b: int) -> List[int]: #k is counter for how many numbers are left to generate, a is the current Fibonacci number, b is the next Fibonacci number
        if k == 0:
            return []
        return [a] + rec(k - 1, b, a + b) #Generate the next Fibonacci number and decrease the counter

    return rec(n, 0, 1)

print(fib_first(10))
```

This function uses a counter variable to limit the output number to the specified ammount of fibonacci numbers. rec is the recursive function that calls itself with the new values.

### B.10.1 Compute the arabic number from a roman numeral, for positive integers from *1* to *3999* . In order to do so use the dict as provided below, if you see fit you may change or add entries:

```python
roman = {I:1, IV: 4, V: 5, IX: 9, X: 10, XL: 40, 
         L: 50, XC: 90, C: 100, CD: 400, D: 500,
         CM:900, M: 1000}
```

```python
ROMANS = [ #This tuple list is ordered from largest to smallest value, names are converted into strings
    ("M", 1000), ("CM", 900), ("D", 500), ("CD", 400),
    ("C", 100),  ("XC", 90),  ("L", 50),  ("XL", 40),
    ("X", 10),   ("IX", 9),   ("V", 5),   ("IV", 4), ("I", 1),
]

def write_roman(n: int) -> str:
    if not (1 <= n <= 3999):
        raise ValueError("n must be between 1 and 3999") 
    out = [] 
    for s, v in ROMANS: # s is the roman numeral string, v is the value
        q, n = divmod(n, v) #q is how many times v fits into n, n is the remainder
        out.append(s * q) #append the roman numeral string q times to the output list
    return "".join(out)

print(write_roman(3724)) #should return "MMMCMXCIX"
print(write_roman(1094)) #should return "IV"
print(write_roman(8)) #should return "XLIV"
```
